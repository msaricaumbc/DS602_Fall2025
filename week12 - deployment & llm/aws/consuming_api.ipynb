```python
import requests
import json
import urllib
```

```python
sentence = "awesome movie but fell asleep"
```

```python
ip = '3.93.181.240/'
sentence_encoded = urllib.parse.quote(sentence)
url = f'http://{ip}/?sentence={sentence_encoded}'
print(url)
```

```python
json.loads(requests.get(url ).text)
```

```python
def predict_sentiment(sentence):
    sentence_encoded = urllib.parse.quote(sentence)
    url = f'http://{ip}/?sentence={sentence_encoded}'
    response = json.loads(requests.get(url ).text)
    prediction = response['sentiment']
    print(prediction)
    return prediction
```

```python
while True:
    sentence = input('>>')
    if not sentence or sentence.strip() == '':
        break

    predict_sentiment(sentence)
```
